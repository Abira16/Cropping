```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

org = cv2.imread('img111.jpg')
pre = cv2.imread('yolo img111.jpg') # 0.20106444529579268,0.16479665440443317

print(org.shape)
dim=(org.shape[1],org.shape[0])
pre_re= cv2.resize(pre, dim)
print(pre_re.shape)
from skimage.metrics import structural_similarity
(score, diff) = structural_similarity(org, pre_re, full=True,multichannel=True)
print("Image similarity", score)
```
